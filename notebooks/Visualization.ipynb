```python
import csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
csv_file = 'data/issues.csv'
prj = []
desc = []
points = []
with open(csv_file, newline='') as file:
    reader = csv.reader(file, quoting=csv.QUOTE_ALL, escapechar='\\', quotechar='"')
    skip = True
    for row in reader:
        if skip:
            skip = False
            continue
        # row = reader[i]
        prj.append(row[0])
        desc.append(row[1])
        points.append(float(row[2]))

max_len = -1
for d in desc:
    if len(d) > max_len:
        if len(d) > 1000:
            print(d)
        max_len = len(d)

print('Max length of description {}'.format(max_len))
print('Project {}; Description {}; Points {}'.format(prj[0], desc[0], points[0]))
print('Projects length = {}'.format(len(prj)))
print('Points length = {}'.format(len(points)))
print("Description len = {}".format(len(desc)))

desc_l = [len(d) for d in desc]
print('l = ', max(desc_l))
```

```python
from matplotlib.patches import Rectangle

plt.figure()
hist = np.histogram(points)
u = np.unique(prj)
desc_l = [len(d) for d in desc]
plt.scatter(points, desc_l, label=u)
plt.show()
```
